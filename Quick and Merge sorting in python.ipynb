```python
def quicksort(arr, left, right):
    # Base case: If the subarray has more than one element
    if left < right:
        # Partition the array and get the partition position
        partition_pos = partition(arr, left, right)
        
        # Recursively apply quicksort on the left and right  subarrays
        quicksort(arr, left, partition_pos - 1)
        quicksort(arr, partition_pos + 1, right)
    
    print(arr, "after sorting") 

def partition(arr, left, right):
    i = left
    j = right - 1
    
    # Select the pivot as the rightmost element
    pivot = arr[right]
    while i < j:
        # Increment i until an element larger than the pivot is found
        while i < right and arr[i] < pivot:
            i += 1
        
        # Decrement j until an element smaller than the pivot is found
        while j > left and arr[j] >= pivot:
            j -= 1
        
        # Swap elements at i and j if they are in the wrong partition
        if i < j:
            arr[i], arr[j] = arr[j], arr[i]
    
    # Place the pivot in its correct position if necessary
    if arr[i] > pivot:
        arr[i], arr[right] = arr[right], arr[i]
    
    return i

arr = [5, 7, 4, 7, 4, 4, 8, 4]
quicksort(arr, 0, len(arr) - 1)
```

```python
import random as r

def mergesort(lst):
    print("this is before sorting", lst)
    
    # Base case: If the list contains more than one element
    if len(lst) > 1:
        left_lst = lst[:len(lst)//2]
        right_lst = lst[len(lst)//2:]
        
        # Recursively apply mergesort on the left half
        mergesort(left_lst)
        
        # Recursively apply mergesort on the right half
        mergesort(right_lst)
        
        i = 0  # Pointer for left list
        j = 0  # Pointer for right list
        k = 0  # Pointer for the merged list
        
        # Merge the left and right halves
        while i < len(left_lst) and j < len(right_lst):
            if left_lst[i] < right_lst[j]:
                lst[k] = left_lst[i]  # Add smaller element from left half
                i += 1
            else:
                lst[k] = right_lst[j]  # Add smaller element from right half
                j += 1
            k += 1
        
        # If there are remaining elements in the left half, add them
        while i < len(left_lst):
            lst[k] = left_lst[i]
            i += 1
            k += 1
        
        # If there are remaining elements in the right half, add them
        while j < len(right_lst):
            lst[k] = right_lst[j]
            j += 1
            k += 1
    
    print("this is after sorting", lst)

def randomlistmaker(L): #makes a list of random elements from 1-100
    n = int(input("enter the length of the list you want"))
    for i in range(n):
        L.append(r.randint(0, 100))
    return L

K = []
mergesort(randomlistmaker(K))
```
